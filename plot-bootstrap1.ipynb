```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#input_filebase = 'correlations-test1c'
#input_filebase = 'correlations-test1f'
#input_filebase = 'correlations-test1f'
#input_filebase = 'correlations-test1f2'
input_filebase = 'correlations/correlations-test1k'
#input_filebase = 'correlations/toygen-correlations-test1c'
```

```python
with open( '%s.npy' % input_filebase, 'rb') as f :
    all_points_df0 = np.load(f)
    all_points_df1 = np.load(f)
    all_points_dr = np.load(f)
    all_points_corr_array = np.load(f)
```

```python
print("\n\n\n")
print("  all_points_df0 shape:  %s" % str(all_points_df0.shape))
print("  all_points_corr_array shape:  %s" % str(all_points_corr_array.shape))
print("\n\n\n")
```

```python
with open( '%s-feature-coords.npy' % input_filebase, 'rb') as f :
    all_points_f0b = np.load(f)
    all_points_f1b = np.load(f)
```

```python
fig,ax = plt.subplots( 1, 3, figsize=(18,5))
ax[0].hist2d( all_points_df0, all_points_corr_array, bins=[80,80], range=([-4,4],[-1,1]))
ax[1].hist2d( all_points_df1, all_points_corr_array, bins=[80,80], range=([-4,4],[-1,1]))
ax[2].hist2d( all_points_dr, all_points_corr_array, bins=[80,80], range=([0,5],[-1,1]))

plt.savefig('correlation-vs-distance.png', bbox_inches='tight')
plt.savefig('correlation-vs-distance.pdf', bbox_inches='tight')

plt.show
```

```python
fig,ax = plt.subplots( 2, 3, figsize=(18,10))

hbins=80
hmax_dr = 8

dr_cut = 1
cor_cut = 0.3

ax[0][0].hist( all_points_corr_array, bins=hbins)
ax[0][0].hist( all_points_corr_array[all_points_dr<dr_cut], bins=hbins, alpha=0.8)

ax[0][1].hist( all_points_corr_array[all_points_dr<dr_cut], bins=hbins)

ax[0][2].hist( all_points_corr_array[all_points_dr>dr_cut], bins=hbins)





ax[1][0].hist( all_points_dr, bins=hbins, range=[0,hmax_dr])
ax[1][0].hist( all_points_dr[all_points_corr_array>cor_cut], bins=hbins, range=[0,hmax_dr], alpha=0.8)

ax[1][1].hist( all_points_dr[all_points_corr_array>cor_cut], bins=hbins, range=[0,hmax_dr])

ax[1][2].hist( all_points_dr[all_points_corr_array<cor_cut], bins=hbins, range=[0,hmax_dr])




plt.show
```

```python
fig,ax = plt.subplots( 1, 2, figsize=(18,9))

hbins=80
hf0min = -4.
hf0max = 4.
hf1min = -6.
hf1max = 6.

cor_cut = 0.8
dr_cut = 0.3

ax[0].hist2d( all_points_f0b, all_points_f1b, 
             bins=[hbins,hbins], range=([hf0min,hf0max],[hf1min,hf1max]) )

ax[1].hist2d( all_points_f0b[(all_points_corr_array>cor_cut)&(all_points_dr>dr_cut)], 
             all_points_f1b[(all_points_corr_array>cor_cut)&(all_points_dr>dr_cut)], 
             bins=[hbins,hbins], range=([hf0min,hf0max],[hf1min,hf1max]) )


plt.show
```
