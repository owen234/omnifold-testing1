```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cov = np.array([[6, -3], [-3, 3.5]])
pts = np.random.multivariate_normal([1, -2], cov, size=80000)
```

```python
pts.mean(axis=0)
```

```python
np.cov(pts.T)
```

```python
np.corrcoef(pts.T)[0, 1]
```

```python

plt.plot(pts[:, 0], pts[:, 1], '.', alpha=0.5)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
plt.hist2d(pts[:, 0], pts[:, 1], bins=100)
plt.axis('equal')
plt.grid()
plt.show()
```
